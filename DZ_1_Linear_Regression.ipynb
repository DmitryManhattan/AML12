```python
#Произведем замему "?" на Nan, чтобы оценить сколько пустых данных у нас есть

adult = adult.replace(['?'],np.nan)
adult.info()
```

```python
#Посольку значения Nan есть только в категориальных атрибутах (т.е. их нельзя заменить на средние значения 
#как если бы речь шла о цифровых показателях), то предлагаю убрать эти данные

adult = adult.dropna()
adult.info()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# выбрать признаки кроме целевой переменной Income
features = adult[['age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country' ]]

# переведем столбцы с категориальными переменными в значения 0 и 1, добавив столбцы с соответствующими названиями
X = pd.get_dummies( features, columns = [ 'workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country' ] )


X.head()

```

```python
# целевая переменная (столбец income) снова является категориальной
# переведем значения столбца в числа, оставив один столбец

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit( adult['income'] )

```

```python
le.classes_
```

```python
y = pd.Series( data = le.transform( adult['income'] ) )
y.head()
```

```python
#Разделим данные на тренировочные и тестовые

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

# Строим Logistic Regression

```python
# обучаем модель
model = LogisticRegression()

model.fit( x_train, y_train )

```

```python
#Смотрим Score какой на тренивочных данных
model.score(x_train, y_train)
```

```python
#Делаем прогноз на основе тестовых данных и смотрим какой score
prediction = model.predict(x_test)
model.score(x_test, y_test)
```
